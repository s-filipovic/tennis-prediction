## 4. MODEL

Modelom logističke regresije opisuje se veza između kategoričke zavisne promenljive i prediktora.

Pretpostavimo da je zavisna promenljiva $Y$ indikator nekog događaja, tj. da ima Bernulijevu raspodelu, čiji parametar zavisi od vrednosti prediktora $X$,

$$Y: \begin{pmatrix}
0 & 1 \\
1-\pi(X) & \pi(X) 
\end{pmatrix}.$$

Potrebno je modelovati verovatnoću $P\{Y=1 | X\} = \pi(X)$, kao $\pi(X) = g(\beta_0+\beta_1X)$, gde je $g$ funkcija koja uzima vrednosti iz $[0,1]$. Postoje razni izbori funkcije $g$. U slučaju logističke regresije koristi se $g(x)=\frac{e^x}{1+e^x}$, što je funkcija raspodele logističke raspodele.

Tada je
$$\pi(X)=\frac{e^{\beta_0+\beta_1X}}{1+e^{\beta_0+\beta_1X}}.$$


Ako je $\pi(X) = g(\beta_0+\beta_1X)$, onda će važiti
$$\log\frac{\pi(X)}{1-\pi(X)} = \beta_0+\beta_1X.$$

Količnik $q(X) = \frac{\pi(X)}{1-\pi(X)}$ se naziva **kvota**. Iz jednakosti $\frac{\pi(X)}{1-\pi(X)} = \frac{P\{Y=1 | X\}}{P\{Y=0 | X\}}$, vidimo da kvota predstavlja odnos verovatnoća da zavisna promenljiva uzme vrednosti $1$ odnosno $0$, za određenu vrednost prediktora.

Parametri modela $\beta$ logističke regresije se ocenjuju metodom maksimalne verodostojnosti, a njihova interpretacija se vidi iz sledećeg.
Kako važi $\frac{q(X+1)}{q(X)}=e^{\beta_1}$, vidimo da znak koeficijenta $\beta_1$ određuje pozitivan ili negativan uticaj prediktora $X$ na verovatnoću $P\{Y=1 | X\}$, a konketna vrednost $e^{\beta_1}$ koliko puta se menja kvota jediničnom promenom prediktora $X$.


U slučaju da imamo više prediktora, model se jednostavno proširuje na
$$\pi(X)=\frac{e^{X\beta}}{1+e^{X\beta}},$$
gde je $\beta$ vektor parametara, a $X$ model matrica.

Logističku regresiju koristićemo za klasifikaciju ishoda teniskog meča, na pobedu ili poraz.

Definišemo pomoćne funkcije getdays i getmonth kojima prosleđujemo datum, a one vraćaju odgovarajući broj dana, odnosno meseci

```python
def getdays(date1):
    year1 = int(date1[0] + date1[1] + date1[2] + date1[3])
    month1 = int(date1[4] + date1[5])
    date1 = int(date1[6] + date1[7])

    answer = 365*year1 + 30*month1 + date1
    return answer

def getmonth(date1):
    month1 = int(date1[4] + date1[5])
    return month1
```

```python
def create_add_year_matches(input_path, output_path):

    with open(input_path, 'r') as infile, open(output_path, 'w') as outfile:
        fieldnames = ['tourney_id',  #0 ID turnira
                      'tourney_name', #1 naziv turnira
                      'surface',  #2 podloga na kojoj se igra
                      'draw_size', #3 veličina žreba, broj igrača na turniru
                      'tourney_level', #4 nivo turnira ('G' grend slem, 'M' masters, 'A' ostali)
                      'tourney_date', #5 datum odigravanja meča
                      'match_num',  #6 ID meča na turniru
                      'winner_id',  #7 ID pobednika
                      'winner_seed', #8 broj koji predstavlja koji je nosilac bio igrač na turniru
                      'winner_entry', #9 način na koji se pobednik kvalifikovao na turnir
                      'winner_name',  #10 ime pobednika
                      'winner_hand',  #11 dominantnija ruka pobednika
                      'winner_ht',    #12 visina pobednika u centimetrima
                      'winner_ioc',   #13 država porekla pobednika
                      'winner_age',   #14 broj godina pobednika
                      'winner_rank',  #15 rang pobednika
                      'winner_rank_points', #16 broj bodova pobednika
                      'loser_id',    #17 ID gubitnika
                      'loser_seed', #18 broj koji predstavlja koji je nosilac bio igrač na turniru
                      'loser_entry', #19 način na koji se gubitnik kvalifikovao na turnir
                      'loser_name', #20 ime gubitnika
                      'loser_hand', #21 dominantnija ruka gubitnika
                      'loser_ht',  #22 visina gubitnika u centimetrima
                      'loser_ioc', #23 država porekla gubitnika
                      'loser_age', #24 broj godina gubitnika
                      'loser_rank', #25 rang gubitnika
                      'loser_rank_points', #26 broj bodova gubitnika
                      'score', #27 rezultat na kraju meča
                      'best_of', #28 broj dobijenih setova do kog se igra
                      'round', #29 faza takmičenja
                      'minutes', #30 koliko min je trajao meč
                      'w_ace', #31 asovi pobednika
                      'w_df', #32 duple grešaka pobednika
                      'w_svpt', #33 broj poena prilikom kojih je pobednik servirao
                      'w_1stIn', #34 broj uspešnih prvih servisa pobednika
                      'w_1stWon',#35 osvojeni poeni pobednika na njegov 1. servis
                      'w_2ndWon', #36 osvojeni poeni pobednika na njegov 2. servis
                      'w_SvGms', #37 broj gemova u kojima je pobednik servirao
                      'w_bpSaved', #38 spašene brejk lopte pobednika
                      'w_bpFaced', #39 ukupan broj brejk lopti gubitnika
                      'l_ace', #40 asovi gubitnika
                      'l_df', #41 duple grešaka gubitnika
                      'l_svpt', #42 broj poena prilikom kojih je gubitnika servirao
                      'l_1stIn', #43 broj uspešnih prvih servisa gubitnika
                      'l_1stWon', #44 osvojeni poeni gubitnika na njegov 1. servis
                      'l_2ndWon', #45 osvojeni poeni gubitnika na njegov 2. servis
                      'l_SvGms', #46 broj gemova u kojima je gubitnik servirao
                      'l_bpSaved', #47 spašene brejk lopte gubitnika
                      'l_bpFaced'] #48 ukupan broj brejk lopti gubitnika

        writer = csv.DictWriter(outfile, fieldnames=fieldnames)
        #menjamo redosled zaglavlja

        writer.writeheader()
        #upisujemo preuređene redove u novi fajl
        for row in csv.DictReader(infile):
            writer.writerow(row)

    inputs = np.loadtxt(output_path, dtype=np.dtype(str), delimiter=',', skiprows=1)
    return inputs
```

**Tournament Form** – ukupan broj pobeda/poraza na određenom turniru

```python
# pravimo 2 rečnika koja se odnose na broj pobeda i ukupan broj odigranih mečeva, 
# gde ključ čine ID igrača i ime turnira, a vrednost ovi brojevi
tournament_form_win = dict()
tournament_form_count = dict()

# prolazimo kroz celu bazu
for i in range(num_rows):
    # uzimamo ID igrača meča
    player_id_winner = inputs[i][7]
    player_id_loser = inputs[i][17]

    # ukoliko pobednik nije igrao na datom turniru ranije, dodajemo ga u rečnik i na početku
    # postavljamo njegov broj pobeda i ukupan broj mečeva na 0
    # ime turnira se nalazi u inputs[i][0][5:] 
    if (player_id_winner, inputs[i][0][5:]) not in tournament_form_win:
        tournament_form_win[(player_id_winner, inputs[i][0][5:])] = 0
        tournament_form_count[(player_id_winner, inputs[i][0][5:])] = 0
    
    # isto radimo i za poraženog igrača
    if (player_id_loser, inputs[i][0][5:]) not in tournament_form_win:
        tournament_form_win[(player_id_loser, inputs[i][0][5:])] = 0
        tournament_form_count[(player_id_loser, inputs[i][0][5:])]  = 0
    
    # povećavamo broj pobeda igrača koji je pobedio, a broj pobeda onog koji je izgubio ostaje isti
    tournament_form_win[(player_id_winner, inputs[i][0][5:])] += 1
    tournament_form_win[(player_id_loser, inputs[i][0][5:])] += 0
    
    # dodajemo ovaj odigrani meč na ukupan broj mečeva oba igrača
    tournament_form_count[(player_id_winner, inputs[i][0][5:])] += 1
    tournament_form_count[(player_id_loser, inputs[i][0][5:])] += 1
    
    # prvo dodajemo vrednosti u trening skup 
    if i < num_rows - count_2022:
        
        # na parna mesta upisujemo vrednosti za pobednika
        # razlika u broju pobeda se nalazi u koloni (44) 
        X_inputs[2*i][44] = tournament_form_win[(player_id_winner, inputs[i][0][5:])] - 1 - tournament_form_win[(player_id_loser, inputs[i][0][5:])]
        # razlika u ukupnom broju odigranih mečeva (45)
        X_inputs[2*i][45] = tournament_form_count[(player_id_winner, inputs[i][0][5:])] - tournament_form_count[(player_id_loser, inputs[i][0][5:])]
        
        # u dve pomoćne promenljive stavljamo ukupan broj odigranih mečeva pobednika i poraženog, redom
        temp11 = 1
        temp12 = 1
        
        if tournament_form_count[(player_id_winner, inputs[i][0][5:])] != 1:
            temp11 = tournament_form_count[(player_id_winner, inputs[i][0][5:])]
        if tournament_form_count[(player_id_loser, inputs[i][0][5:])]  != 1:
            temp12 = tournament_form_count[(player_id_loser, inputs[i][0][5:])] 
        
        # sada računamo razliku u broju pobeda izraženu u procentima (46)
        X_inputs[2*i][46] = ((tournament_form_win[(player_id_winner, inputs[i][0][5:])] - 1)*100/temp11) - ((tournament_form_win[(player_id_loser, inputs[i][0][5:])])*100/temp12)
        
        # poraženi igrač će imati iste vrednosti, samo suprotnog znaka
        # njih upisujemo na neparna mesta u bazi
        X_inputs[2*i+1][44] = -X_inputs[2*i][44]
        X_inputs[2*i+1][45] = -X_inputs[2*i][45]
        X_inputs[2*i+1][46] = -X_inputs[2*i][46]
        
    # unosimo vrednosti u test skup
    else:
        # one će se u bazi nalaziti nakon svih vrednosti za trening skup
        x1 = int(2*i - 2*int(num_rows) + 2*int(count_2022))
        
        # kao i za trening skup unosimo vrednosti za razliku u pobedama, ukupan broj mečeva i 
        # razliku u broju pobeda izraženu u procentima
        # prvo unosimo za pobednika
        X_prediction[x1][44] = tournament_form_win[(player_id_winner, inputs[i][0][5:])] - 1 - tournament_form_win[(player_id_loser, inputs[i][0][5:])]
        X_prediction[x1][45] = tournament_form_count[(player_id_winner, inputs[i][0][5:])] - tournament_form_count[(player_id_loser, inputs[i][0][5:])]

        temp11 = 1
        temp12 = 1

        if tournament_form_count[(player_id_winner, inputs[i][0][5:])] != 1:
            temp11 = tournament_form_count[(player_id_winner, inputs[i][0][5:])]
        if tournament_form_count[(player_id_loser, inputs[i][0][5:])]  != 1:
            temp12 = tournament_form_count[(player_id_loser, inputs[i][0][5:])] 

        X_prediction[x1][46] = ((tournament_form_win[(player_id_winner, inputs[i][0][5:])] - 1)*100/temp11) - ((tournament_form_win[(player_id_loser, inputs[i][0][5:])])*100/temp12)
        
        # a zatim i za gubitnika
        X_prediction[x1+1][44] =  -X_prediction[x1][44]
        X_prediction[x1+1][45] = -X_prediction[x1][45]
        X_prediction[x1+1][46] = -X_prediction[x1][46]
```

**Match based Form** - broj pobeda/poraza u poslednjih 5/10/15/25 mečeva

```python
# pravimo rečnik za broj pobeda/poraza u poslednjih 5/10/15/25 mečeva,
# gde je ključ ID igrača, a vrednost uređeni par (0/1, datum), 
# gde 0 predstavlja poraz, a 1 pobedu
form = dict()

# isto kao ranije, prolazimo kroz celu bazu i uzimamo ID pobednika i gubitnika
for i in range(num_rows):
    player_id_winner = inputs[i][7]
    player_id_loser = inputs[i][17]
    
    # ukoliko pobednik nije u bazi dodajemo ga i upisujemo njegovu pobedu, zajedno sa njenim datumom 
    if player_id_winner not in form:
        form[player_id_winner] = []
        form[player_id_winner].append((1, inputs[i][5]))
    
    # slično postupamo i za poraženog, ovaj put upisujući poraz
    if player_id_loser not in form:
        form[player_id_loser] = []
        form[player_id_loser].append((0, inputs[i][5]))
    
    # u sledeće promenljive upisujemo ukupan broj odigranih mečeva pobednika pre posmatranog meča
    # zbog toga ih inicijalizujemo na -1
    total_winner_5 = -1
    total_winner_10 = -1
    total_winner_15 = -1
    total_winner_25 = -1
    
    # na isti način postupamo za gubitnika
    total_loser_5 = -1
    total_loser_10 = -1
    total_loser_15 = -1
    total_loser_25 = -1

    # broj pobeda i pobednika i gubitnika inicijalizujemo na 0
    winner_win_5 = 0
    winner_win_10 = 0
    winner_win_15 = 0
    winner_win_25 = 0

    loser_win_5 = 0
    loser_win_10 = 0
    loser_win_15 = 0
    loser_win_25 = 0

    # kako želimo ishode poslednjih 5/10/15/25 mečeva, kroz rečnik prolazimo od kraja
    # pravimo po 4 for ciklusa, za svaki od ovih slučajeva i redom za pobednika i gubitnika
    for (a1, a2) in reversed(form[player_id_winner]):
        # u slučaju da je ukupan broj mečeva pred trenutni -1, on postaje 0
        if total_winner_5 == -1:
            total_winner_5 = 0
        else:
            # ukoliko još nismo izbrojali 5 mečeva, na broj pobeda u poslednjih 5 mečeva pobednika
            # dodajemo trenutnu i povećavamo ukupan broj mečeva
            if total_winner_5 < 5:
                winner_win_5 = winner_win_5 + a1
                total_winner_5 += 1
    
    # analogno radimo ostale slučajeve
    for (a1, a2) in reversed(form[player_id_winner]):
        if total_winner_10 == -1:
            total_winner_10 = 0
        else:
            if total_winner_10 < 10:
                winner_win_10 = winner_win_10 + a1
                total_winner_10 += 1

    for (a1, a2) in reversed(form[player_id_winner]):
        if total_winner_15 == -1:
            total_winner_15 = 0
        else:
            if total_winner_15 < 15:
                winner_win_15 = winner_win_15 + a1
                total_winner_15 += 1

    for (a1, a2) in reversed(form[player_id_winner]):
        if total_winner_25 == -1:
            total_winner_25 = 0
        else:
            if total_winner_25 < 25:
                winner_win_25 = winner_win_25 + a1
                total_winner_25 += 1

    # sada ponavljamo postupak za gubitnika
    for (a1, a2) in reversed(form[player_id_loser]):
        if total_loser_5 == -1:
            total_loser_5 = 0
        else:
            if total_loser_5 < 5:
                loser_win_5 = loser_win_5 + a1
                total_loser_5 += 1

    for (a1, a2) in reversed(form[player_id_loser]):
        if total_loser_10 == -1:
            total_loser_10 = 0
        else:
            if total_loser_10 < 10:
                loser_win_10 = loser_win_10 + a1
                total_loser_10 += 1

    for (a1, a2) in reversed(form[player_id_loser]):
        if total_loser_15 == -1:
            total_loser_15 = 0
        else:
            if total_loser_15 < 15:
                loser_win_15 = loser_win_15 + a1
                total_loser_15 += 1

    for (a1, a2) in reversed(form[player_id_loser]):
        if total_loser_25 == -1:
            total_loser_25 = 0
        else:
            if total_loser_25 < 25:
                loser_win_25 = loser_win_25 + a1
                total_loser_25 += 1

    # u trening skup upisujemo razlike dobijenih vrednosti, u kolone 36-39 
    if i < num_rows - count_2022:
        X_inputs[2*i][36] = winner_win_5 - loser_win_5
        X_inputs[2*i][37] = winner_win_10 - loser_win_10
        X_inputs[2*i][38] = winner_win_15 - loser_win_15
        X_inputs[2*i][39] = winner_win_25 - loser_win_25

        X_inputs[2*i+1][36] = loser_win_5 - winner_win_5
        X_inputs[2*i+1][37] = loser_win_10 - winner_win_10
        X_inputs[2*i+1][38] = loser_win_15 - winner_win_15
        X_inputs[2*i+1][39] = loser_win_25 - winner_win_25
    
    # takođe i u test skup
    else:
        x1 = int(2*i - 2*int(num_rows) + 2*int(count_2022))
        
        X_prediction[x1][36] = winner_win_5 - loser_win_5
        X_prediction[x1][37] = winner_win_10 - loser_win_10
        X_prediction[x1][38] = winner_win_15 - loser_win_15
        X_prediction[x1][39] = winner_win_25 - loser_win_25

        X_prediction[x1+1][36] = loser_win_5 - winner_win_5
        X_prediction[x1+1][37] = loser_win_10 - winner_win_10
        X_prediction[x1+1][38] = loser_win_15 - winner_win_15
        X_prediction[x1+1][39] = loser_win_25 - winner_win_25
```

**Time based Form** - broj pobeda/poraza u poslednjih 30 dana

```python
# ponovo koristimo rečnik form

for i in range(num_rows):
    player_id_winner = inputs[i][7]
    player_id_loser = inputs[i][17]
    
    # želimo da postupimo na sličan način kao u prethodnom koraku, dakle inicijalizujemo 
    # promenljive koje se odnose na broj pobeda i ukupan broj mečeva za oba igrača i 
    # for ciklusima prolazimo kroz bazu, kretajući sa kraja
    total_winner_1 = -1
    total_loser_1 = -1
    winner_win_1 = 0
    loser_win_1 = 0

    for (a1, a2) in reversed(form[player_id_winner]):
        if total_winner_1 == -1:
            total_winner_1 = 0
        else:
            # zanima nas slučaj u kom je meč odigran u poslednjih 30 dana, a dalje postupamo kao u prethodnom koraku
            if getdays(inputs[i][5]) - getdays(a2) < 30 and getdays(inputs[i][5]) - getdays(a2) >= 0:
                winner_win_1 = winner_win_1 + a1
                total_winner_1 += 1
            else:
                break

    for (a1, a2) in reversed(form[player_id_loser]):
        if total_loser_1 == -1:
            total_loser_1 = 0
        else:
            if getdays(inputs[i][5]) - getdays(a2) < 30 and getdays(inputs[i][5]) - getdays(a2) >= 0:
                loser_win_1 = loser_win_1 + a1
                total_loser_1 += 1
            else:
                break

    # odgovarajuće vrednosti unosimo u trening i test skup
    if i < num_rows - count_2022:
        X_inputs[2*i][47] = winner_win_1 - loser_win_1
        X_inputs[2*i+1][47] = loser_win_1 - winner_win_1
        X_inputs[2*i][48] = total_winner_1 - total_loser_1
        X_inputs[2*i+1][48] = total_loser_1 - total_winner_1

    else:
        x1 = int(2*i - 2*int(num_rows) + 2*int(count_2022))
        
        X_prediction[x1][47] = winner_win_1 - loser_win_1
        X_prediction[x1+1][47] = loser_win_1 - winner_win_1
        X_prediction[x1][48] = total_winner_1 - total_loser_1
        X_prediction[x1+1][48] = total_loser_1 - total_winner_1
```

```python
# pravimo listu rečnika gde svaki rečnik ima ključ player_id i vrednost statistiku

player_id_stats_overall_sum = [dict() for x in range(num_cols)]

# ovo će nam poslužiti kao brojač za računanje proseka kasnije
player_id_stats_overall_count = [dict() for x in range(num_cols)]

career_stats_winner = np.zeros((num_rows, num_cols))
career_stats_winner_total = np.zeros((num_rows, num_cols))
career_stats_loser = np.zeros((num_rows, num_cols))
career_stats_loser_total = np.zeros((num_rows, num_cols))

for i in range(num_rows):
    player_id_winner = inputs[i][7]
    player_id_loser = inputs[i][17]
    
    # j je brojač koji prolazi kroz statistike
    for j in range(31, 40):
         # j je ovakav jer u rečniku brojimo od 0, a u inputs statistike su tek od 32. kolone i idu do 40. za pobednika
         # pa kao i do sad, inicijalizacija igrača ako ga nema a zatim i beleženje odgovarajućih vrednosti
        if player_id_winner not in player_id_stats_overall_count[j-31]:
            career_stats_winner[i][j] = 0
            career_stats_winner_total[i][j] = 0

            player_id_stats_overall_count[j-31][player_id_winner] = 1
            player_id_stats_overall_sum[j-31][player_id_winner] = int(inputs[i][j])
        else:
            career_stats_winner[i][j] = player_id_stats_overall_sum[j-31][player_id_winner]/player_id_stats_overall_count[j-31][player_id_winner]
            career_stats_winner_total[i][j] = player_id_stats_overall_sum[j-31][player_id_winner]

            player_id_stats_overall_count[j-31][player_id_winner] = player_id_stats_overall_count[j-31][player_id_winner] + 1 
            player_id_stats_overall_sum[j-31][player_id_winner] = player_id_stats_overall_sum[j-31][player_id_winner] + int(inputs[i][j])
         
    for j in range(40, 49):
         # j uzima ove vrednsoti jer u ovom rečniku brojimo od 0, a u inputs statistike su tek od 41. kolone i idu do 49. za gubitnika
        if player_id_loser not in player_id_stats_overall_count[j-40]:
            career_stats_loser[i][j] = 0
            career_stats_loser_total[i][j] = 0

            player_id_stats_overall_count[j-40][player_id_loser] = 1
            player_id_stats_overall_sum[j-40][player_id_loser] = int(inputs[i][j])
        else:
            career_stats_loser[i][j] = player_id_stats_overall_sum[j-40][player_id_loser]/player_id_stats_overall_count[j-40][player_id_loser]
            career_stats_loser_total[i][j] = player_id_stats_overall_sum[j-40][player_id_loser]

            player_id_stats_overall_count[j-40][player_id_loser] = player_id_stats_overall_count[j-40][player_id_loser] + 1 
            player_id_stats_overall_sum[j-40][player_id_loser] = player_id_stats_overall_sum[j-40][player_id_loser] + int(inputs[i][j])

        # i kao i do sad završavamo sa popunjavanjem test i trening skupa
        if i < num_rows - count_2022:
            x1 = int(2*i)
            
            X_inputs[x1][j-40] = career_stats_winner[i][j-9] - career_stats_loser[i][j]
            X_inputs[x1][j-24] = career_stats_winner_total[i][j-9] - career_stats_loser_total[i][j]
            X_inputs[x1+1][j-40] = -X_inputs[x1][j-40]
            X_inputs[x1+1][j-24] = -X_inputs[x1][j-24]

        else:
            x1 = int(2*i - 2*int(num_rows) + 2*int(count_2022))
            
            X_prediction[x1][j-40] = career_stats_winner[i][j-9] - career_stats_loser[i][j]
            X_prediction[x1][j-24] = career_stats_winner_total[i][j-9] - career_stats_loser_total[i][j]
            X_prediction[x1+1][j-40] = -X_prediction[x1][j-40]
            X_prediction[x1+1][j-24] = -X_prediction[x1][j-24]
```

Potrebno je još da postavimo vrednosti koje se predviđaju, odnosno da u svakom parnom redu upisujemo jedinice (tj. pobede), a u svakom neparnom upisujemo nule (odnosno poraze).

Za klasifikaciju ishoda teniskog meča koristićemo logističku regresiju, gde ćemo svakom igraču dodeliti ili klasu 0 (poražen) ili klasu 1 (pobednik). Koristimo funkciju LogisticRegression koja se nalazi u modulu sklearn.linear_model iz paketa scikit-learn. 

Parametri koje zadajemo ovoj funkciji su:<br>
multi_class='ovr'  - u pitanju je binarna klasifikacija<br>
random_state=0     - slično kao i set.seed<br>
solver='liblinear' - algoritam koji se koristi u optimizacionim problemima<br>
penalty='l2'       - označava da koristimo L2 regularizaciju

Clf je klasifikator, koji u sebi sadrži predict_proba u kojima se nalaze verovatnoće da igrač bude pobednik (odnosno da pripada klasi 1) i da igrač bude poražen (odnosno da pripada klasi 0)

```python
clf = LogisticRegression(multi_class='ovr', random_state=0, solver='liblinear', penalty='l2').fit(X_inputs, Y_inputs)
# beležimo verovatnoće pripadnosti klasama
training_prediction = clf.predict_proba(X_prediction)
```

Ukoliko označimo verovatnoće sa:<br>
    verovatnoća da prvi igrač izgubi - L1<br>
    verovatnoća da prvi igrač pobedi - W1<br>
    verovatnoća da drugi igrač izgubi - L2<br>
    verovatnoća da drugi igrač pobedi - W2<br>
    
Posmatrajući verovatnoće pripadnosti klasama koje smo zabeležili u training_prediction, razlikujemo dva slučaja:<br>
    Prvi slučaj: Verovatnoće su u skladu jedna sa drugom:<br>
    Npr. ako je L1 < W1 (verovatnije je da je prvi igrač pobedio) i<br>
         ako je L2 > W2(verovatnije je da je drugi igrač izgubio)<br>
Ove verovatnoće su u skladu jedna sa drugom, odnosno ima smisla predvideti pobedu prvog igrača<br>

Analogno za drugog igrača:<br>
    Npr. ako je L1 > W1 (verovatnije je da je prvi igrač izgubio) i<br>
         ako je L2 < W2(verovatnije je da je drugi igrač pobedio)<br> 
Ove verovatnoće su u skladu jedna sa drugom, odnosno ima smisla predvideti pobedu drugog igrača (tj. poraz prvog)<br>

Drugi slučaj: Verovatnoće nisu u skladu jedna sa drugom:<br>
    Tada sabiramo verovatnoće saglasnih izjava i onda poredimo njih odnosno<br>
    Poredimo L1 + W2 (oba se slažu sa porazom prvog) i W1 + L2 (oba se slažu sa pobedom prvog)<br>
Predviđamo onaj ishod koji ima veću verovatnoću od ta dva zbira<br>

```python
total = count_2022
right = 0

# petlja po svim mečevima iz 2022. godine
for i in range(count_2022):
    (L1,W1) = training_prediction[2*i]
    (L2,W2) = training_prediction[2*i+1]

    if L1 < W1 and L2 > W2:
        predicted = 1
    else: 
        if L1 > W1 and L2 < W2:
            predicted = 0
        else:
            if L1 + W2 < W1 + L2:
                predicted = 1
            else:
                predicted = 0

    # proveravamo da li je model dobro predvideo tj. poredimo sa tačnim vrednostima 
    if Y_prediction[2*i] == predicted:
        right = right + 1
        
print("Od ukupno " + str(total) + " mečeva, model je tačno predvideo pobednika " + str(right) + " puta.")
print("Tačnost je " + str(right*100/total) + "%.")
```
